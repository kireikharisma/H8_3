#### Hapus Kolom
Bersihkan dataset dengan menghapus beberapa kolom yang tidak perlu. Hal ini bisa dilakukan dengan metode pandas drop()

```python
# kolom AREA, REG, DEV, Type, Coverage tidak diperlukan sehingga kolom tersebut dapat dihapus/dihilangkan
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True) # axis=1 : dua dimensi
df_can.head(2) 
```

#### Ubah Nama Kolom
Mengganti nama kolom dengan metode rename() dan memasukkan dictionary nama lama dan nama baru

```python
# untuk baris yang ada null dihapus
df_can.dropna(axis=0, inplace=True)
df_can.tail(2)
```

```python
df_can.drop(df_can.tail(2).index, inplace=True) # drop last n rows
```

```python
df_can.info()
```

```python
# mengganti nama kolom dengan replace(), nama_lama : nama_baru
df_can.rename(columns={'OdName' : 'Country', 'AreaName' : 'Continent', 'RegName' : 'Region'}, inplace=True)
df_can.columns
```

#### Menambahkan dan Menjumlahkan Kolom
Meenambahkan kolom 'Total' yang merangkum total imigran menurut negara selama periode 1980-20113

```python
df_can['Total'] = df_can.sum(axis=1, numeric_only=True) # horizontal/columns, axis=0 : vertical/row
df_can.head()
```

```python
# tipe data dionvert menjadi integer
df_can['Total'] = df_can['Total'].astype('Int64')
df_can.head()
```

```python
# grouping berdasarkan region kemudian dijumlahkan
print(df_can.groupby('Region').sum())

# atau bisa juga grouping berdasarkan kolom
#df_can.groupby(['col_1', 'col_2'], as_index=False).mean()
#print(df_can.groupby('Continent').sum())
```
