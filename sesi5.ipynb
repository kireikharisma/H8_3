## Combining Multiple Dataset

```python
further_city_data = pd.DataFrame(
    {"Revenue" : [7000, 3400], "employee_count": [2,2]},
    index = ["New York", "Barcelona"]
)
```

```python
further_city_data
```

```python
city_data
```

```python
#menggabungkan dataset:

all_city_data = pd.concat([city_data, further_city_data], sort=False)
```

```python
all_city_data
```

```python
city_countries = pd.DataFrame({
    "country": ["Holland", "Japan", "Holland", "Canada", "Spain"],
    "capital": [1,1,0,0,0]},
    index= ["Amsterdam", "Tokyo", "Rotterdam", "Toronto", "Barcelona"])
```

```python
city_countries
```

```python
all_city_data
```

```python
#concat mampu menambahkan baris (axis = 0) dan kolom (axis = 1)
cities = pd.concat([all_city_data, city_countries], axis=1, sort=False)
```

```python
cities
```

```python
#menggabungkan kota yg hanya muncul di kedua objek DataFrame:

pd.concat([all_city_data, city_countries], axis=1, join="inner")
```

```python
#menyatukan beberapa dataset menggunakan .merge:

countries = pd.DataFrame ({
    "population_millions": [17, 127, 37],
    "continent": ["Europe", "Asia", "North America"]
}, index = ["Holland", "Japan", "Canada"])
```

```python
countries
```

```python
cities
```

```python
#merge melakukan inner join scr default

pd.merge(cities, countries, left_on="country", right_index=True)
```

```python
#jika ingin menampilkan semua nama kota harus menggunakan how:

pd.merge(cities, countries, left_on="country", right_index=True,
        how="left")
```

#### Membuat Plot

Beberapa argumen untuk membuat jenis plot yg berbeda dengan pandas:
- 'line': line plot (default)
- 'bar' : vertical bar plot
- 'barh' : horizontal bar plot
- 'hist' : histogram
- 'box' : boxplot
- 'kde' : kernel density estimation plot
- 'density' : same as 'kde'
- 'area' : area plot
- 'pie' : pie plot
- 'scatter' : scatter plot
- 'hexbin' : hexbin plot

```python
nba["fran_id"].value_counts().head(10).plot(kind="line")
```
