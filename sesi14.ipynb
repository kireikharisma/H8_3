# Principal Component Analysis
case: menggunakan The Breast Cancer data set, terdiri dari 2 kelas (malignant dan benign). kelas malignant memiliki 212 sampel dan kelas benign memiliki 357 sampel. dataset ini memiliki 30 fitur yang dibagikan di semua kelas: radius, texture, perimeter, area, smoothness, fractal dimension, etc.

**Data Exploration**

```python
#import modul
from sklearn.datasets import load_breast_cancer
```

```python
# .data untuk mengambil data, .target untuk mengambil label
breast = load_breast_cancer()
breas_data = breast.data
print(breas_data.shape)

breast_labels = breast.target
print(breast_labels.shape)
```

```python
#import numpy untuk membentuk kembali breast_label agar bisa membuat dataframe berisi data dan label
import numpy as np

labels = np.reshape(breast_labels,(569,1)) #reshaping label
```

setelah reshaping label, kita akan concatenate data dan label di sepanjang second axis, yang berarti bentuk akhir dari array akan menjadi 569 x 31

```python
final_breast_data = np.concatenate([breas_data,labels],axis=1)
final_breast_data.shape
```

```python
#import pandas untuk membuat dataframe
import pandas as pd
```

```python
breast_dataset = pd.DataFrame(final_breast_data)
```

```python
breast_dataset
```

```python
#print the features
features = breast.feature_names
features
```

jika dilihat dalam array fitur, bidang label hilang. oleh karena itu, kita harus menambahkan secara manual ke array fitur karena kita akan menyamakan array ini dengan nama kolom dari dataframe breast_dataset kita

```python
features_labels = np.append(features,'label')
```

```python
#menyematkan nam akolom ke dataframe breast_dataset
breast_dataset.columns = features_labels
breast_dataset.head()
```

mengubah label menjadi benign dan malignant menggunakan fungsi .replace dan inplace = true yang akan memodifikasi dataframe breast_dataset

```python
breast_dataset['label'].replace(0, 'Benign', inplace=True)
breast_dataset['label'].replace(1, 'Malignant', inplace=True)
```

```python
breast_dataset.tail()
```

# Data Visualization using PCA

**visualizing the breast cancer data**

standardizing data karen akeluaran PCA dipengaruhi berdasarkan skala fitur data. ini adalah praktik umum untuk menormalkan data kita sebelum memasukkannya ke machine learning algorithm apapun.\
 \
untuk menerapkan normalization, kita akan mengimpor modul StandardScaler dari sklearn library dan hanya memilih fitur dari breast_dataset yang kita buat di langkah Eksplorasi Data. setelah kita memiliki fiturnya, kita kemudian akan menerapkan scalling dengan melakukan fit_transform pada data fitur.\
 \
saat menerapkan StandardScaler, setiap fitur data kita harus didistribusikan secara normalsehinga akan menskalakan distribusi ke rata-rata nol dan deviasi standar satu.

```python
from sklearn.preprocessing import StandardScaler
x = breast_dataset.loc[:, features].values
x = StandardScaler().fit_transform(x) #normalizing the features
```

```python
x.shape
```

ubah fitur yang dinormalkan menjadi format tabel dengan bantuan DataFrame.

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_breast = pd.DataFrame(x, columns=feat_cols)
normalised_breast.tail()
```

```python
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
```

buat dataframe yang akan memiliki PCA untuk semua 569 sampel

```python
principal_breast_Df = pd.DataFrame(data=principalComponents_breast,
                                  columns=['Principal Component 1',
                                          'Principal Component 2'])
```

```python
principal_breast_Df.tail()
```

```python
#nilai varians setiap PCA
print('Explained variation per principal component:', format(pca_breast.explained_variance_ratio_))
```

dari output dapat diamati bahwa PCA 1 menampung 44,2% informasi sedangkan PCA 2 hanya menampung 19% informasi. juga, hal ini yang perlu diperhatikan adalah saat memproyeksikan data 30 dimensi menjadi data dua dimensi, 36,8% informasi hilang

**visualisasi**

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1', fontsize=20)
plt.ylabel('Principal Component - 2', fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset", fontsize=20)
targets=['Benign', 'Malignant']
colors=['r','g']
for target, color in zip (targets, colors):
    indicesToKeep = breast_dataset['label'] ==target
    plt.scatter(principal_breast_Df.loc[indicesToKeep, 'Principal Component 1'],
               principal_breast_Df.loc[indicesToKeep, 'Principal Component 2'], c=color, s=50)
plt.legend(targets,prop={'size':15})
plt.show()
```

dari grafik diatas dapat diamati bahwa dua kelas benign dan malignant, ketika diproyeksikan ke ruang dua dimensi, dapat dipisahkan secara linier sampai batar tertentu. pengamatan lain dapat berupa penyebaran kelas benign dibandingkan dengan kelas malignant.

# K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
X = -2 * np.random.rand(100,2)
X1 = 1 + 2 * np.random.rand(50,2)
X[50:100, :] = X1

plt.scatter(X[:, 0], X[:, 1], s=50, c='b')
plt.show()
```

```python
Kmean = KMeans(n_clusters=2)
Kmean.fit(X)
```

```python
KMeans(algorithm='auto', copy_x=True, init='k-means++', max_iter=300,
      n_clusters=2, n_init=10, n_jobs=None,
      precompute_distance='auto',
      random_state=None, tol=0.0001, verbose=0)
```

```python
Kmean.cluster_centers_
```

# Finalize Your Model with Pickle

```python
import pandas as pd
```

```python
#Assigning features and label variables
weather = ['Sunny', 'Sunny', 'Overcast', 'Rainy', 'Rainy', 'Rainy', 'Overcast', 'Sunny', 'Sunny', 'Rainy', 'Sunny', 'Overcast', 'Overcast', 'Rainy']
temp = ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 'Mild', 'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild']
play = ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No']

#Import LabelEncoder
from sklearn import preprocessing

#creating labelEncoder
le = preprocessing.LabelEncoder()

#converting string labels into numbers.
weather_encoded = le.fit_transform(weather)

#converting string labels into numbers
temp_enconded = le.fit_transform(temp)
label = le.fit_transform(play)

#combining weather and temp into single listof tuples
features = list(zip(weather_encoded, temp_enconded))

#import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#create a Gaussian Classifier
model = GaussianNB()

#train the model using the training sets
model.fit(features, label)

#save the model to disk
import pickle

filename = 'finalized_model.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
loaded_model = pickle.load(open('finalized_model.sav', 'rb'))
```
