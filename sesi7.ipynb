#### Diagram Pie/Pie Chart

Diagram pie : menyajikan data menggunakan lingkaran dan biasanya menampilkan persentase data yang menginformasikan proporsi objek suatu data terhadap keseluruhan data

function : pie()
function pie() tidak memiliki sumbu dan hanya memiliki satu parameter x yang merepresentasikan nilai data yang diketahui, pie(x)


Beberapa parameter : 
- autopct : menampilkan nilai persentase

#### Scatter Plot/Scatter Diagram

Scatter plot : melihat sebaran data antara dua variabel untuk mendapatkan/mengetahui hubungan atau korelasi antara dua variabel tersebut

Korelasi antara dua variabel dapat berupa korelasi positif, korelasi negatif, ataupun tidak ada korelasinya

#### Manggabungkan Beberapa Grafik Plot dalam Satu Figure

Beberapa grafik plot dapat digabungkan dalam satu figure dengan menggunakan function subplot(m, n, i)
- parameter m dan n : dimensi dari grafik plot
- parameter i : urutan dari grafik plot

```python
x = np.array([1, 2, 3, 4, 5, 6])
y0 = x ** 2
y1 = x ** 3

#plt.subplot(m, n, i) # m = 1 dan n = 2,  1 baris dengan 2 kolom
plt.subplot(1, 2, 1)        # untuk plot pertama, i = 1
plt.plot(x, y0, "r-")
plt.title("Judul Plot Y0")
plt.xlabel("Sumbu x")
plt.ylabel("Sumbu y")
plt.legend(['Y0=x^2'])

plt.subplot(1, 2, 2)        # untuk plot kedua, i = 2
plt.plot(x, y1, "r--")
plt.title("Judul Plot Y1")
plt.xlabel("Sumbu x")
plt.ylabel("Sumbu y")
plt.legend(['Y1=x^3'])

plt.suptitle("2 Grafik Plot")
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5, 6])
y0 = x ** 2
y1 = x ** 3
y2 = x ** 4

#plt.subplot(m, n, i) # m = 3 dan n = 2,  3 baris dengan 2 kolom
plt.subplot(3, 1, 1)        # untuk plot pertama, i = 1
plt.plot(x, y0, "r-")
plt.title("Judul Plot Y0")
plt.xlabel("Sumbu x")
plt.ylabel("Sumbu y")
plt.legend(['Y0=x^2'])

plt.subplot(3, 1, 2)        # untuk plot kedua, i = 2
plt.plot(x, y1, "r--")
plt.title("Judul Plot Y1")
plt.xlabel("Sumbu x")
plt.ylabel("Sumbu y")
plt.legend(['Y1=x^3'])

plt.subplot(3, 1, 3)        # untuk plot ketiga, i = 3
plt.plot(x, y1, "ro")
plt.title("Judul Plot Y3")
plt.xlabel("Sumbu x")
plt.ylabel("Sumbu y")
plt.legend(['Y2=x^4'])

plt.suptitle("3 Grafik Plot")
plt.show()
```

#### Area Plot

Di bagian terakhir, kita membuat plot garis yang memvisualisasikan 5 negara teratas yang menyumbang paling banyak imigran ke Kanada dari 1980-2013

Dengan sedikit modifikasi pada kode, kita dapat mmemvisualisasikan plot ini dengan cumulative plot, juga dikenal dengan Stacked Line Plot atau Area Plot

```python
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)

#get the top 5 entries
df_top5 = df_can.head()

#transpose the daataframe
df_top5 = df_top5[years].transpose()

df_top5.head()
```

Area plot ter-stacked secara default. Dan untuk menhasilkan area plot,

#### Mengconvert ke dalam file image

```python
import base64
from io import BytesIO

from flask import Flask
from matplotlib.figure import Figure

app = Flask(__name__)

@app.route("/")
def hello() :
    # Generate the figure without using pyplot
    fig = Figure()
    ax = fig.subplots()
    ax.plot([1,2])
    # Save it to a temporary buffer
    buf = BytesIO()
    fig.savefig(buf, format="png")
    # Embed the result in the html output
    data = base64.b64encode(buf.getbuffer()).decode("ascii")
    return f"<img src='data:image/png:base64, {data}'/>"
```
